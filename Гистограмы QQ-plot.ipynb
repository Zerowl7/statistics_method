```python
import numpy as np

import pylab 

import matplotlib.pyplot as plt

import scipy.stats as stats

 

mean = 0  # . 

sd = 50    #  

 

#   QQ-plot

sample = np.random.normal(loc=mean, scale=sd, size=20)  #     . 100,   50      1000

plt.figure(figsize = [6.4*2, 4.8])  #   qq_plot

plt.subplot(1, 2, 1)

stats.probplot(sample, dist="norm", plot=plt)  #   QQ-plot         

plt.title('QQ-plot / -')  # 

plt.xlabel(' Z-')  #    X

plt.ylabel('  ')  #    Y

plt.grid(True)  #  

 

#  

plt.subplot(1, 2, 2)

n, bins, patches = plt.hist(sample, 5, density=True, facecolor='g', alpha=0.75, label='')  # C 

plt.title(' ')  # 

plt.xlabel(' ')  #    X

plt.ylabel(' ')  #    Y

plt.text(-3, 0.35, r'$\mu=0,\ \sigma=1$')  # 

plt.grid(True)  #  

 

#     Z-

x = np.linspace(mean - 3*sd, mean + 3*sd, 500)  #    Z-

plt.plot(x, stats.norm.pdf(x, mean, sd), color='red', label='Z-')  #   Z-

 

plt.legend()  #  

plt.show()  #  
```
